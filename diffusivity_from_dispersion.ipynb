#Calculating Diffusivity from Dispersion

##TRACMASS Diffusivity Tuning to Match Relative Dispersion
Drifters that originate at the same time and location will follow the same path unless some sort of subgrid scale parameterization is included. In Doos et al 2011 (*Dispersion of surface drifters and model-simulated trajectories*), a comparison of relative dispersion was made between real and modeled surface drifters. The horizontal diffusivity parameter input to TRACMASS was then adjusted in order to match the drifter relative dispersion at the end of 32 days.

##Gulf of Mexico Calculations of Relative Dispersion from Drifter Data
LaCasce et al 2003 (*Relative dispersion at the surface of the Gulf of Mexico*) use SCULP surface drifters tracks to calculate many metrics, one of which was relative dispersion. They found that this metric did not depend much on position, though the boundary current along the Texas coast does bias the dispersion in that area.

<img src="files/figures/lacasce_dispersion.png" alt="Relative Dispersion" height="500" width="500">

<img src="files/figures/lacasce_sculp1.png" alt="Drifter set 1" height="500" width="500">
<img src="files/figures/lacasce_sculp2.png" alt="Drifter set 2" height="500" width="500">

LaCasce et al define the relative dispersion as:
$D^2(t)= 1/N \Sigma_{i\ne j} (x_i(t)-x_j(t))^2 + (y_i(t)-y_j(t))^2$, and the initial separation distance of the pairs of drifters was less than or equal to 1 km, where $N$ is the number of drifter pairs.

* LaCasce found little difference in the drifter dispersion whether over shallow or deeper water, but there was anisotropic introduced for drifters along the western Gulf shore, which was attributed to the boundary current.
* The first 10 days of dispersion is approximately linear on this log scale plot, so there is exponential growth, up to 40-50km separation.
* From days 10 to 50, there is a power law relation to the drifter dispersion.

##TracPy Dispersion Simulations

I ran a simulation of drifters every 700 meters in projected x and y coordinates (max distance between drifters 990 meters) seeded in the box bounded by [-95,-94,27.5,28.5], and ran them forward for 25 days without any subgrid scale parameterization.

```python
# Lists of simulation runs
dispersion_root = 'dispersion_calculations/'
tracks_root = 'tracks'
runsA = list(('_doturb0', '_doturb1_ah5', '_doturb1_ah20', '_doturb1_ah100', 
             '_doturb2_ah5', '_doturb2_ah20', '_doturb3_ah20'))
runsB = list(('B_doturb0', 'B_doturb1_ah20', 'B_doturb1_ah50', 'B_doturb2_ah20', 'B_doturb1_ah35', 'B_doturb2_ah10', 'B_doturb3_ah5', 'B_doturb2_ah5'))
runsC = list(('C_doturb0', 'C_doturb1_ah50', 'C_doturb2_ah20', 'C_doturb1_ah35', 'C_doturb2_ah10', 'C_doturb1_ah20', 'C_doturb2_ah5'))
# LaCasce Data
lacasce = np.loadtxt('figures/lacasce_dispersion_Points.txt')
```

```python
colors = ['0', '.7', '.3'] # no diff, doturb=1, doturb=2,3
symbols = ['-', '--', '-.'] 

fig = figure(figsize=(22,6))
subplot(121)
plt.semilogy(lacasce[:,0], lacasce[:,1], 'rx', label='LaCasce', markersize=7)
plt.xlabel('Days'); plt.ylabel('Relative Dispersion (km$^2$)');
ax = gca()
text(0.025, 0.01, 'Data', color='r', transform=ax.transAxes)
text(0.025, 0.93, 'No diffusion', color='k', transform=ax.transAxes)
text(0.25, 0.93, 'Turbulent fluctuations (--), A$_H= 5, 20, 100 m^2/s$', color='.6', transform=ax.transAxes)
text(0.2, 0.01, 'Random displacement (:), A$_H= 5, 20 m^2/s$', color='.4', transform=ax.transAxes)
title('Seeding in area A')
for run in runsA:
    Dname = os.path.join(dispersion_root, 'D2' + run + '.npz')
    if not os.path.exists(Dname):
        D2, t = dispersion.calc_dispersion(os.path.join(tracks_root, '2006-02-01T00' + run + '.nc'), grid)
        np.savez(Dname, D2=D2, t=t)
    else:
        D = np.load(Dname)
        D2 = D['D2']; t = D['t']; D.close()
    days = (t-t[0])/(3600.*24) # time vector in days
    doturb = int(run.split('doturb')[1][0]) # this gives the doturb value for plotting
    if doturb == 3: doturb=2;
    plt.semilogy(days, D2, symbols[doturb], color=colors[doturb], linewidth=2)
    
subplot(122)
plt.semilogy(lacasce[:,0], lacasce[:,1], 'rx', label='LaCasce', markersize=7)
plt.xlabel('Days'); plt.ylabel('Relative Dispersion (km$^2$)');
ax = gca()
text(0.025, 0.01, 'Data', color='r', transform=ax.transAxes)
text(0.025, 0.93, 'No diffusion', color='k', transform=ax.transAxes)
text(0.25, 0.93, 'Turbulent fluctuations (--), A$_H= 20, 35, 50 m^2/s$', color='.6', transform=ax.transAxes)
text(0.2, 0.01, 'Random displacement (:), A$_H= 5, 10, 20 m^2/s$', color='.4', transform=ax.transAxes)
title('Seeding in area B')
for run in runsB:
    Dname = os.path.join(dispersion_root, 'D2' + run + '.npz')
    if not os.path.exists(Dname):
        D2, t = dispersion.calc_dispersion(os.path.join(tracks_root, '2006-02-01T00' + run + '.nc'), grid)
        np.savez(Dname, D2=D2, t=t)
    else:
        D = np.load(Dname)
        D2 = D['D2']; t = D['t']; D.close()
    days = (t-t[0])/(3600.*24) # time vector in days
    doturb = int(run.split('doturb')[1][0]) # this gives the doturb value for plotting
    if doturb == 3: doturb=2;
    plt.semilogy(days, D2, symbols[doturb], color=colors[doturb], linewidth=2)

fig = figure(figsize=(22,6))
subplot(121)
plt.semilogy(lacasce[:,0], lacasce[:,1], 'rx', label='LaCasce', markersize=7)
plt.xlabel('Days'); plt.ylabel('Relative Dispersion (km$^2$)');
ax = gca()
text(0.025, 0.01, 'Data', color='r', transform=ax.transAxes)
text(0.025, 0.93, 'No diffusion', color='k', transform=ax.transAxes)
text(0.25, 0.93, 'Turbulent fluctuations (--), A$_H= 20, 35,50 m^2/s$', color='.6', transform=ax.transAxes)
text(0.2, 0.01, 'Random displacement (:), A$_H= 5,10,20 m^2/s$', color='.4', transform=ax.transAxes)
title('Seeding in area C')
for run in runsC:
    Dname = os.path.join(dispersion_root, 'D2' + run + '.npz')
    if not os.path.exists(Dname):
        D2, t = dispersion.calc_dispersion(os.path.join(tracks_root, '2006-02-01T00' + run + '.nc'), grid)
        np.savez(Dname, D2=D2, t=t)
    else:
        D = np.load(Dname)
        D2 = D['D2']; t = D['t']; D.close()
    days = (t-t[0])/(3600.*24) # time vector in days
    doturb = int(run.split('doturb')[1][0]) # this gives the doturb value for plotting
    if doturb == 3: doturb=2;
    plt.semilogy(days, D2, symbols[doturb], color=colors[doturb], linewidth=2)
```

Based on a discussion with Rob, it makes the most sense to use the behavior of the numerical drifter relative dispersion for up to the first week or so. This is because a shorter time scale (1 to a few days) would be the time scale over which the model is missing additional energy due to unresolved small scales.

There is different behavior depending on where the drifters were seeded, the time, and which subgrid parameterization is being used. Relative dispersion results in Doos were relatively similar between the added turbulence and diffusion schemes, but were much too high using the turbulence scheme. So, I will use the diffusion scheme. Based on the results above, and trying to match the relative dispersion in the first 7 days, I will use the random displacement scheme on a circle with a horizontal diffusivity of $A_H=5 m^2/s$. And I am not using the elliptical version of the random diffusion because it gives basically the same results as the circle one.

<img src="files/figures/doos_absdispersion.png" alt="Absolute Dispersion" height="500" width="500">


```python
colors = ['.6', '.3', '.6', '.3', '.6', '.3']
symbols = ['--', '-', '--', '-', '--', '-']
alphas = [.5, 1, .5, 1, .5, 1]
runs = list(('_doturb0', '_doturb2_ah5', 'B_doturb0', 'B_doturb2_ah5', 'C_doturb0', 'C_doturb2_ah5'))

fig = figure(figsize=(11,6))
plt.semilogy(np.array([7,7]), np.array([10**(-1), 10**(5)]), ':', color='lightgrey', linewidth=3)
for i, run in enumerate(runs):
    Dname = os.path.join(dispersion_root, 'D2' + run + '.npz')
    D = np.load(Dname)
    D2 = D['D2']; t = D['t']; D.close()
    plt.semilogy(days, D2, symbols[i], color=colors[i], linewidth=2, alpha=alphas[i])
plt.semilogy(lacasce[:,0], lacasce[:,1], 'ro', label='LaCasce', markersize=7)
plt.xlabel('Days'); plt.ylabel('Relative Dispersion (km$^2$)');
ax = gca()
text(0.025, 0.015, 'Data (o)', color='r', transform=ax.transAxes)
text(0.17, 0.015, 'No diffusion (--)', color='.6', transform=ax.transAxes)
text(0.415, 0.015, 'Random displacements (-)', color='.3', transform=ax.transAxes)
plt.savefig('figures/diffusivity.pdf', bbox_inches='tight')
```

The above figure shows the relative dispersion for numerical drifters with no diffusion and with diffusion added via random displacements on a circle, with a horizontal diffusivity of $A_H=5m^2/s$. The three lines are for three different seeding starting areas. The focus of this adjustment is on the 0 to 7 day time scale (marked in (:)).
