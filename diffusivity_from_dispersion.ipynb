#Calculating Diffusivity from Dispersion

##TRACMASS Diffusivity Tuning to Match Relative Dispersion
Drifters that originate at the same time and location will follow the same path unless some sort of subgrid scale parameterization is included. This parameterization acts to capture some of the subgrid scale energy that is missing in an inherent discretely-resolved numerical model. In Doos et al 2011 (*Dispersion of surface drifters and model-simulated trajectories*), a comparison of relative dispersion was made between real and modeled surface drifters. The horizontal diffusivity parameter input to TRACMASS was then adjusted in order to match the drifter relative dispersion at the end of 32 days.

##Gulf of Mexico Calculations of Relative Dispersion from Drifter Data
LaCasce et al 2003 (*Relative dispersion at the surface of the Gulf of Mexico*) use SCULP surface drifters tracks to calculate many metrics, one of which was relative dispersion. They found that this metric did not depend much on position, though the boundary current along the Texas coast does bias the dispersion in that area. They use drifters from both the SCULP 1 and 2 surveys, but we will use only SCULP 1 because those drifters originated on the Texas-Louisiana shelf, which is within the TXLA domain. The drifter starting locations are shown below. For our simulations, the drifters are seeded in the same approximate box as the SCULP 1 drifters.

<img src="files/figures/lacasce_dispersion.png" alt="Relative Dispersion" height="500" width="500">
<img src="files/figures/driftersloc_dimarco2001.png" alt="Drifter locations" height="500" width="500">

LaCasce et al define the relative dispersion as:
$D^2(t)= 1/N \Sigma_{i\ne j} (x_i(t)-x_j(t))^2 + (y_i(t)-y_j(t))^2$, and the initial separation distance of the pairs of drifters was less than or equal to 1 km, where $N$ is the number of drifter pairs.

* LaCasce found little difference in the drifter dispersion whether over shallow or deeper water, but there was anisotropic introduced for drifters along the western Gulf shore, which was attributed to the boundary current.
* The first 10 days of dispersion had approximately exponential growth with growth rate 0.55.
* From days 10 to 50, there is a power law relation to the drifter dispersion, with power 2.2 according to relative dispersion or up to 3 by other metrics.


##TracPy Dispersion Simulations

I ran a simulation of drifters every 700 meters in projected x and y coordinates (diagonal distance between drifters 990 meters) seeded in the box bounded by [-93.8,-92.2,28,29.2], and ran them forward for 50 days without any subgrid scale parameterization, and with combinations of types of diffusion and different parameter values. To copy the timing of the drifters as closely as possible, I started the drifters weekly for October, November, and December; biweekly for January, February, and March; and monthly for April, May, June, and July (LaCasce 2003). I ran these for 2009-2010 in order to overlap with the CRCM model output and the Deepwater Horizon oil spill.

Differences in setups:

* LaCasce used any pairs that happened to be near each other and most pairs were chance pairs. In my setup, all pairs are taken that are initialized near each other, since the experiment was designed to evaluate pair behavior.
* different years
* I use way more pairs

Based on a discussion with Rob, it makes the most sense to use the behavior of the numerical drifter relative dispersion for up to the first week or so. This is because a shorter time scale (1 to a few days) would be the time scale over which the model is missing additional energy due to unresolved small scales.

There is different behavior depending on where the drifters were seeded, the time, and which subgrid parameterization is being used. Relative dispersion results in Doos were relatively similar between the added turbulence and diffusion schemes, but were much too high using the turbulence scheme. So, I will use the diffusion scheme. Based on the results above, and trying to match the relative dispersion in the first 7 days, I will use the random displacement scheme on a circle with a horizontal diffusivity of $A_H=5 m^2/s$. And I am not using the elliptical version of the random diffusion because it gives basically the same results as the circle one.
